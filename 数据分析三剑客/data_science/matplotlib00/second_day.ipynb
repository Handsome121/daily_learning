# 柱状图绘制

```python
movie_name = ['雷神','正义联盟','全球风暴','降魔传','追捕']
tickets = [77777,88888,88888,9999,5555]

plt.figure(figsize=(20,8),dpi = 80)

x_tickets = range(len(movie_name))

plt.bar(x_tickets,tickets,color=['b','r','g','y','k'])

plt.xticks(x_tickets,movie_name)

plt.title('电影票房收入对比')

plt.grid(True,linestyle = '--',alpha = 0.5)

plt.xlabel('电影名称',fontsize=16)
plt.ylabel('票房数',fontsize=16)

plt.show()
```

```python
movie_name = ['雷神','正义联盟','全球风暴','降魔传','追捕']
first_tickets = [77777,88888,88888,9999,5555]
first_weekend = [8888,9999,5555,6666,9999]

plt.figure(figsize=(20,8),dpi = 80)

x_tickets = range(len(movie_name))
plt.bar(x_tickets,first_tickets,color='b',width=0.2,label='首日票房')
plt.bar([i+0.2 for i in x_tickets],first_weekend,width=0.2,label='首周票房')

plt.xticks([i+0.1 for i in x_tickets],movie_name)

plt.legend()

plt.show()
```
